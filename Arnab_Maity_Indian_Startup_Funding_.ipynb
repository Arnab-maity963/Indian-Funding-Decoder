```python
# SECTION A :

# Q1.
# Series : It is a one-dimensional object that stores a single column of data.
# DataFrame : It is two-dimensional and contains multiple columns and rows.
# Both support operations like .head() but operations such as .merge() or .groupby() are mainly used with DataFrames.

# Q2.

# inplace=True : It  changes the original DataFrame directly instead of creating a new one.
# Pandas is discouraging its use because it can make code harder to understand and doesn't always improve performance.
#  For example, using df = df.drop('Age', axis=1) is usually clearer than df.drop('Age', axis=1, inplace=True).

# Q3.

# .fillna() : It replaces missing values with a specified value.
# .dropna() : It removes rows or columns containing missing values.
# A data analyst may use .fillna(0) for missing sales data but use .dropna() when rows with missing values are incomplete and not useful for analysis.

# Q4.

# errors ='coerce' converts invalid values into NaN instead of raising an error.
# In the  real-world messy datasets, that option is preferred because it allows data cleaning without stopping the program when non-numeric values are present in th dataset.

# Q5.

# .loc : It selects data using row or column labels.
# .iloc : It selects data using integer positions. If a DataFrame index starts from 100 instead of 0.
# df.loc[0] looks for the label 0, whereas df.iloc[0] returns the first row, so they may produce different results.

# Q6.

# .map() : used to transform each value in a Series, usually with a dictionary or function.
# apply() : Used to runs a custom function on each value, while .replace() substitutes specific values without applying any function.
# Each is useful depending on whether you need mapping, custom processing, or simple replacement.

# Q7.

# SettingWithCopyWarning appears when you try to modify a copy of a DataFrame instead of the original.
# For example, df[df['Age']>18]['Status']='Adult' triggers the warning. It can be fixed by using .loc, such as df.loc[df['Age']>18, 'Status']='Adult'.

# Q8.

# pd.cut() : It divides data into fixed value ranges and pd.qcut() creates groups with an equal number of observations.
# For annual income labels like Low, Mid, and High, pd.qcut() is  better because it balance the number of customers in each category, making comparisons more meaningful.
```
